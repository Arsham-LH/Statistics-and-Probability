```python
import numpy as np
from scipy.stats import bernoulli as ber
import matplotlib.pyplot as plt
from scipy.stats import poisson
def poissonLim(n,total,landa):
    X=np.array([[0]*total]*n)
    Y=np.zeros(total)
    for i in range(n): #from 0 to n-1
        X[i]=np.array([ber.rvs(landa/n,size=total)])
        Y=Y+X[i]
    return Y

landa=5
total=10000 #total samples
n1,n2,n3,n4,n5=np.array([10,100,1000,10000,100000])

#plotting poisson distribution PMF
fig=plt.figure(figsize=(10,6))
fig.suptitle('poisson PMF')
maxVal=20 #maximum value to show on plot
r=range(maxVal)
plt.subplot(111) #plotting PMF of poisson distribution
Y=poisson(mu=landa) #Y~poisson(landa)
plt.bar(r,Y.pmf(r)) #discrete plot for PMF
plt.ylabel('pY(r)')
plt.xlabel('r')
plt.grid(True)
plt.show()
```

```python
#plotting histogram for n=n1=10, using the function defined in previous cell
Y1=poissonLim(n1,total,landa)
max1=20 #maximum value to show on plot

fig = plt.figure(figsize=(10,6)) #creating figure
fig.suptitle('Y1 normalized histogram') #figure title
plt.subplot(111) #normalized histogram

plt.hist(Y1,range=(0,max1),bins=max1,rwidth=1,density=True) #values range from 0 to 20. every bin width is set to 1. density=True, makes the histogram normalized 
plt.ylabel('normalized freq')
plt.xlabel('Y1')
plt.grid(True)
```
