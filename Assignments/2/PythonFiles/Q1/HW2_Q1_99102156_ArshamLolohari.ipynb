```python
from scipy.stats import bernoulli as ber
from scipy.stats import binom
from scipy.stats import nbinom
import numpy as np
import matplotlib.pyplot as plt
p=0.4
total=1000 #total samples
total_vars=10 #total Xi variables
x1,x2,x3,x4,x5,x6,x7,x8,x9,x10=np.array([ber.rvs(p,size=total),ber.rvs(p,size=total),ber.rvs(p,size=total),ber.rvs(p,size=total),ber.rvs(p,size=total),ber.rvs(p,size=total),ber.rvs(p,size=total),ber.rvs(p,size=total),ber.rvs(p,size=total),ber.rvs(p,size=total)])
Y1=x1+x2+x3+x4+x5+x6+x7+x8+x9+x10 #Y1=x1+x2+...+x10
fig = plt.figure(figsize=(15,6)) #creating figure
fig.suptitle('Y1 normalized histogram & PMF') #figure title
plt.subplot(121) #first plot: normalized histogram

plt.hist(Y1,range=(0,total_vars),bins=10,rwidth=1,density=True) #Y1 values range from 0 to 10. every bin width is set to 1. density=True, makes the normalized histogram
plt.ylabel('normalized freq')
plt.xlabel('Y1')
plt.grid(True)

plt.subplot(122) #second plot: PMF of binomial distribution
r=range(total_vars+1) #random variable Range={0,1,...,10}
Y1_2=binom(total_vars,p) #Y1_2~Binomial(10,0.4)
plt.bar(r,Y1_2.pmf(r)) #discrete plot for PMF
plt.ylabel('pY(r)')
plt.xlabel('r')
plt.grid(True)
plt.show()
```

```python
from numpy.random import geometric as geo
p2=0.6
total=1000
total_vars2=3
z1,z2,z3=np.array([geo(p2,total),geo(p2,total),geo(p2,total)])
Y2=z1+z2+z3

max_num=18 #maximum horizontal value on plots
fig = plt.figure(figsize=(15,6)) #creating figure
fig.suptitle('Y2 normalized histogram & PMF') #figure title
plt.subplot(121) #first plot: normalized histogram

plt.hist(Y2,bins=max_num,range=(0,max_num),rwidth=1,density=True) #Y2 values range from 0 to 18. there are 18 bins. every bin width is set to 1. density=True, makes the normalized histogram
plt.ylabel('normalized freq')
plt.xlabel('Y2')
plt.grid(True)

plt.subplot(122) #second plot: PMF of pascal(negative binomial) distribution
Y2_2=nbinom(total_vars2,p2,loc=total_vars2) #Y1_2~pascal(3,0.6), loc=3 for 3units shift
losses=range(max_num) #number of losses to reach 3 successes
plt.bar(losses,Y2_2.pmf(losses)) #discrete plot for PMF
plt.ylabel('pY(r2)') #r2: number of experiments to reach 3 successes = losses+3
plt.xlabel('r2')
plt.grid(True)
plt.show()
```
