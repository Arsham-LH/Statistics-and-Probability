```python
import numpy as np
from numpy.random import geometric as geo
from numpy.random import poisson
from numpy.random import binomial as binom
import matplotlib.pyplot as plt

N=10000 #total samples
p1,p3=0.4,0.1 #p for X1,X3
n1=5 #n for X1
landa=1.6 #landa for X2 (poisson)
X1=binom(n1,p1,size=N)
X2=poisson(landa,size=N)
X3=geo(p3,size=N)
Y=X1+X2+X3

maxVal=80 #maximum value to show on plot

fig = plt.figure(figsize=(10,6)) #creating figure
fig.suptitle('Y normalized histogram') #figure title
plt.subplot(111) #normalized histogram

plt.hist(Y,range=(0,maxVal),bins=maxVal,rwidth=1,density=True) #values range from 0 to 80. every bin width is set to 1. density=True, makes the histogram normalized 
plt.ylabel('normalized freq')
plt.xlabel('Y')
plt.grid(True)

mean=Y.sum()/N
Var=(((Y-mean)**2).sum())/N

print(mean)
print(Var)
```

```python
from scipy.stats import bernoulli as ber
def createY(n,total,p):
    X=np.array([[0]*total]*n)
    Y=np.zeros(total)
    for i in range(n): #from 0 to n-1
        X[i]=np.array([ber.rvs(p,size=total)])
        Y=Y+X[i]
    return Y/n



total=1000 #total samples
n1,n2,n3=5,40,300
p=0.8
maxVal=1 #maximum value to show on plots
Y1=createY(n1,total,p)
Y2=createY(n2,total,p)
Y3=createY(n3,total,p)

fig = plt.figure(figsize=(15,6))
fig.suptitle('Y1&Y2&Y3 normalized histograms')

plt.subplot(131)
plt.hist(Y1,range=(0,maxVal),bins=50,rwidth=1,density=True) #values range from 0 to 1. every bin width is set to 0.02. density=True, makes the normalized histogram
plt.ylabel('normalized freq')
plt.xlabel('Y1')
plt.grid(True)

plt.subplot(132)
plt.hist(Y2,range=(0,maxVal),bins=50,rwidth=1,density=True) #values range from 0 to 1. every bin width is set to 0.02. density=True, makes the normalized histogram
plt.ylabel('normalized freq')
plt.xlabel('Y2')
plt.grid(True)

plt.subplot(133)
plt.hist(Y3,range=(0,maxVal),bins=50,rwidth=1,density=True) #values range from 0 to 1. every bin width is set to 0.02. density=True, makes the normalized histogram
plt.ylabel('normalized freq')
plt.xlabel('Y3')
plt.grid(True)

mean1=Y1.sum()/total
mean2=Y2.sum()/total
mean3=Y3.sum()/total

Var1=(((Y1-mean1)**2).sum())/total
Var2=(((Y2-mean2)**2).sum())/total
Var3=(((Y3-mean3)**2).sum())/total

print([mean1,mean2,mean3])
print([Var1,Var2,Var3])
```
