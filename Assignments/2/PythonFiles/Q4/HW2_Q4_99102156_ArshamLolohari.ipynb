```python
import numpy as np
from random import randint
import matplotlib.pyplot as plt


total=10000 #total experiments
totDays=365 #total days of the year
def exp(): # a function to do a single experiment
    days=[] #written days
    i=randint(1,totDays) #new random day from 1 to 365
    stages=0 #total steps so far
    while (not(i in days)): #while i is a new day
        stages=stages+1 #one more step
        days.append(i) #add day i to written days
        i=randint(1,totDays) #new random value for i
    return stages

Y=np.zeros(total) 
prob = [0]*totDays
max_prob = [0]*totDays
for i in range(total): #setting experiments results
    Y[i]=exp()
for i in range(totDays): #calculating prob and max_prob
    prob[i]=(list(Y).count(i+1))/total
    max_prob[i]=sum(prob)

maxVal=100 #maximum value to show on plots

fig = plt.figure(figsize=(16,6)) #creating figure
fig.suptitle('Y histogram & probability') #figure title
plt.subplot(131) #histogram

plt.hist(Y,range=(0,maxVal),bins=maxVal,rwidth=1) #values range from 0 to 100. every bin width is set to 1.
plt.ylabel('freq')
plt.xlabel('Y')
plt.grid(True)

plt.subplot(132) #prob plot
day=range(1,totDays+1)
plt.bar(day,prob)
plt.xlim(1,100)
plt.ylabel('probability')
plt.xlabel('n(guests)')
plt.grid(True)
print('prob= ',prob)

nForHalfProb=1
for i in range(totDays):
    if max_prob[i]<0.5:
        i=i+1
        nForHalfProb=nForHalfProb+1
print('minimum n for half prob= ',nForHalfProb)

plt.subplot(133) #max_prob plot
plt.bar(day,max_prob)
plt.xlim(1,100)
plt.ylabel('probability for maximum n guests')
plt.xlabel('n')
plt.grid(True)
```
