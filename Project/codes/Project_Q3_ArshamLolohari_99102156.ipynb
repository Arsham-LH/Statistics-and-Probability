```python
import numpy as np
from scipy.stats import bernoulli as ber
from numpy.linalg import eig
import matplotlib.pyplot as plt

m=10 #number of rows
n=500 #number of columns
X=np.zeros((m,n))
X_all=[]
for i in range(m):
    X[i]=np.random.normal(loc=0,scale=1,size=n) #creating n standard normal random numbers for each row
    X_all=np.concatenate((X_all,X[i]),axis=None) #for plotting all elements distribution

#plotting rows distribution
fig = plt.figure(figsize=(15,8)) #creating figure
fig.suptitle('Rows distribution histogram') #figure title
maxVal=5 #maximum value around 0 to show in the plots
binsNum=50 #total number of bins

for i in range(m):
    plt.subplot(2,5,(i+1)) #normalized histogram

    plt.hist(X[i],range=(-maxVal,maxVal),bins=binsNum,density=True) #values range from -maxVal to maxVal. density=True, makes the histogram normalized 
    plt.title('Row'+str(i+1)) #setting title for each plot
    plt.grid(True)

#plotting all elements distribution
fig2 = plt.figure(figsize=(10,6)) #creating figure
fig2.suptitle('all elements distribution histogram') #figure title
maxVal=5 #maximum value around 0 to show in the plots
binsNum=50 #total number of bins

plt.hist(X_all,range=(-maxVal,maxVal),bins=binsNum,density=True) #values range from -maxVal to maxVal. density=True, makes the histogram normalized 
plt.grid(True)


```

```python
import numpy as np
from scipy.stats import bernoulli as ber
from numpy.linalg import eig
import matplotlib.pyplot as plt

start=20 #starting number of rows and cols
end=29 #ending number of rows and cols
total_mats=(end-start+1)**2 #total number of matrices with different rows/cols
plotsPerRow=5 #number of plots per row

m=start #number of rows
n=start #number of cols

#for plotting all elements distribution
fig = plt.figure(figsize=(15,50)) #creating figure
fig.suptitle('all elements distribution histogram') #figure title
maxVal=5 #maximum value around 0 to show in the plots
binsNum=75 #total number of bins

X=np.zeros((m,n))
X_all=[]
for i in range(start,end+1): #for rows
    for j in range(start,end+1): #for cols
        m=i #updating number of rows 
        n=j #updating number of cols
        X_all=[] #clearing X_all
        X=np.zeros((m,n)) #clearing X 
        
        for k in range(m): #creating X_all, row by row
            X[k]=np.random.normal(loc=0,scale=1,size=n) #creating n standard normal random numbers for each row
            X_all=np.concatenate((X_all,X[k]),axis=None) #for plotting all elements distribution
            
        plt.subplot(int(total_mats/plotsPerRow),plotsPerRow,(i-start)*plotsPerRow*2+(j-start)+1) #new subplot
        plt.hist(X_all,range=(-maxVal,maxVal),bins=binsNum, density=True) #values range from -maxVal to maxVal. density=True, makes the histogram normalized 
        plt.title(str(m)+'*'+str(n)) #setting title for each plot
        plt.grid(True)



```
